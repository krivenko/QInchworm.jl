# Analytic chain diagram evaluation

Benchmark quasi-Monte Carlo integration of high-order strong coupling diagrams by studying a single topology with variable expansion order.

Here the diagrams are computed semi analytically for the chain of connected hybridization lines $\Delta$.

Using the one-level expression

$$
\Delta(\tau) = V^2 \frac{e^{-\epsilon \tau}}{1 + e^{-\beta \epsilon}}
$$

for the hybridization structure $\Delta(\tau)$.

All strong coupling diagrams has a *backbone* of pseudo particle propagators $P(\tau)$. Here we adopt the simple model

$$
P(\tau) = e^{- \eta \tau}
$$

with $\eta = (\ln 2) / \beta$ giving $P(0) = 1$ and $P(\beta) = 1/2$ (corresponding to a single fermion state with zero energy).


Since $P(\tau)$ is a simple exponential (unitary time evolution), it fulfills the product structure $P(\tau) \cdot P(\tau') = P(\tau - \tau')$. This reduces all the chained backbone products to $P(\tau)$

$$
\begin{array}{l}
P(\beta) = 
P(\beta - 0) \\
P(\beta) = 
P(\beta - \tau_1) P(\tau_1 - \tau_2) P(\tau_2 - 0) \\
P(\beta) = 
P(\beta - \tau_1) P(\tau_1 - \tau_2) P(\tau_2 - \tau_3) P(\tau_3 - \tau_4) P(\tau_4 - 0)
\end{array}
$$

so in the strong coupling diagrams $P$ only enters as a prefactor $P(\tau)$.

## Diagram structure

The $n$:th order single connected chain diagram $D_n(\tau)$ have the structure

$$
D_1 = \int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 \, 
[ P(\beta - \tau_1) P(\tau_1 - \tau_2) P(\tau_2 - 0) ]
\\
\Delta(\tau_1 - \tau_2)
$$

$$
D_2 = \int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 \int_0^{\tau_2} d\tau_3 \int_0^{\tau_3} d\tau_4\, 
[ P(\beta - \tau_1) P(\tau_1 - \tau_2) P(\tau_2 - \tau_3) P(\tau_3 - \tau_4) P(\tau_4 - 0) ]
\\
\Delta(\tau_1 - \tau_3) \Delta(\tau_2 - \tau_4) 
$$

$$
D_3 = 
\int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 
\int_0^{\tau_2} d\tau_3 \int_0^{\tau_3} d\tau_4
\int_0^{\tau_4} d\tau_5 \int_0^{\tau_5} d\tau_6
\,
[ P(\beta - \tau_1) P(\tau_1 - \tau_2) P(\tau_2 - \tau_3) P(\tau_3 - \tau_4) P(\tau_4 - \tau_5) P(\tau_5 - \tau_6) P(\tau_6 - 0) ]
\\
\Delta(\tau_1 - \tau_3) \Delta(\tau_2 - \tau_5) \Delta(\tau_4 - \tau_6)
$$

$$
D_4 = 
\int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 
\int_0^{\tau_2} d\tau_3 \int_0^{\tau_3} d\tau_4
\int_0^{\tau_4} d\tau_5 \int_0^{\tau_5} d\tau_6
\int_0^{\tau_6} d\tau_7 \int_0^{\tau_7} d\tau_8
\,
[ P(\beta - \tau_1) P(\tau_1 - \tau_2) P(\tau_2 - \tau_3) P(\tau_3 - \tau_4) P(\tau_4 - \tau_5) P(\tau_5 - \tau_6) P(\tau_6 - \tau_7) P(\tau_7 - \tau_8) P(\tau_8 - 0) ]
\\
\Delta(\tau_1 - \tau_3) \Delta(\tau_2 - \tau_5) \Delta(\tau_4 - \tau_7) \Delta(\tau_6 - \tau_8)
$$

Using the product property of $P$ gives

$$
D_1 = P(\beta) \int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 \, 
\Delta(\tau_1 - \tau_2)
$$

$$
D_2 = P(\beta) \int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 \int_0^{\tau_2} d\tau_3 \int_0^{\tau_3} d\tau_4\, 
\Delta(\tau_1 - \tau_3) \Delta(\tau_2 - \tau_4) 
$$

$$
D_3 = P(\beta) \int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 \int_0^{\tau_2} d\tau_3 \int_0^{\tau_3} d\tau_4\, 
\int_0^{\tau_4} d\tau_5 \int_0^{\tau_5} d\tau_6\, 
\Delta(\tau_1 - \tau_3) \Delta(\tau_2 - \tau_5) \Delta(\tau_4 - \tau_6)
$$

$$
D_4 = 
P(\beta)
\int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 
\int_0^{\tau_2} d\tau_3 \int_0^{\tau_3} d\tau_4
\int_0^{\tau_4} d\tau_5 \int_0^{\tau_5} d\tau_6
\int_0^{\tau_6} d\tau_7 \int_0^{\tau_7} d\tau_8
\\
\Delta(\tau_1 - \tau_3) \Delta(\tau_2 - \tau_5) \Delta(\tau_4 - \tau_7) \Delta(\tau_6 - \tau_8)
$$


### Variable separation

Separating variables in $\Delta$ gives

$$
\Delta(\tau - \tau') = \delta e^{-\epsilon \tau} e^{+\epsilon \tau'}
$$

with $\delta =  V^2 /(1 + e^{-\beta \epsilon})$. Inserting this now gives the structure

$$
D_1 = \delta P(\beta) 
\int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 
\,
e^{-\epsilon \tau_1} e^{+\epsilon \tau_2}
$$

$$
D_2 = \delta^2 P(\beta) 
\int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 
\int_0^{\tau_2} d\tau_3 \int_0^{\tau_3} d\tau_4 
\,
e^{-\epsilon \tau_1}
e^{-\epsilon \tau_2}
e^{+\epsilon \tau_3}
e^{+\epsilon \tau_4}
$$

$$
D_3 = \delta^3 P(\beta) 
\int_0^\beta d\tau_1 \int_0^{\tau_1} d\tau_2 
\int_0^{\tau_2} d\tau_3 \int_0^{\tau_3} d\tau_4
\int_0^{\tau_4} d\tau_5 \int_0^{\tau_5} d\tau_6
\,
e^{-\epsilon \tau_1}
e^{-\epsilon \tau_2}
e^{+\epsilon \tau_3}
e^{-\epsilon \tau_4}
e^{+\epsilon \tau_5}
e^{+\epsilon \tau_6}
$$

$$
D_n = \delta^n P(\beta) 
\iint_{\beta > \tau_1 > ... > 0} d\vec{\tau} \,\,
e^{-\epsilon \tau_1}
\left[ \prod_{k=2}^{2n-1} e^{(-1)^{k-1} \epsilon \tau_k} \right] e^{+\epsilon \tau_{2n}} 
$$

The evaluation can be performed iteratively

$$
I_1(\tau) = \int_0^\tau \, d\tau' \, e^{+\epsilon \tau'}
$$

$$
I_n(\tau) = \int_0^\tau \, d\tau' e^{-\epsilon \tau'} \int_0^{\tau'} d\tau'' \,
e^{+ \epsilon \tau''} I_{n-1}(\tau'')
$$

$$
D_n = \delta^n P(\beta) \int_0^\beta d\tau \, e^{-\epsilon \tau} I_n(\tau)
$$

# Sympy implementation

Straight forward evaluation using the iterative formula in Sympy.

Reasonable speed for orders $\lesssim 100$

```python
def evaluate_diagrams_slow(n_max, subs=dict(), digits=15):
    """ Compute D_n for all n = 1, ..., n_{max} using Sympy """
    
    t, tp = sp.symbols(rf"\tau, \tau'", real=True)
    ϵ, δ, β = sp.symbols(r"\epsilon, \delta, \beta", real=True, nonzero=True)

    if subs[ϵ] == 0:
        ϵ = 0

    def get_D(I, n, subs):
        D = δ**n / 2 * sp.integrate(sp.exp(-ϵ*t) * I, (t, 0, β))
        D = D.subs(subs)
        D_num = D.n(n=digits)
        return D_num
    
    I1 = sp.integrate(sp.exp(ϵ*tp), (tp, 0, t))
    D1 = get_D(I1, 1, subs)
    
    Ds = [D1]
    
    In = I1
    for n in tqdm(range(2, n_max+1)):
        In = sp.integrate(sp.exp( ϵ*t ) * In, (t,  0, tp)).expand()
        In = sp.integrate(sp.exp(-ϵ*tp) * In, (tp, 0, t )).expand()
        Dn = get_D(In, n, subs)
        Ds.append(Dn)
            
    Ds = np.array(Ds, dtype=float)
    return Ds
```

# Specialized evaluation

Using the closed form expression

$$
\int x^n e^{ax} dx = \frac{e^{ax}}{a^{n+1}} \sum_{k=0}^n (-1)^{n-k} (n-k)! (ax)^k = e^{ax} P_{n,\alpha}(x)
$$

and a function call cache to avoid re-evaluation of the $P_{n,\alpha}(x)$ polynomials, in combination with a custom class to represent sums of single exponents and polynomial products

$$
f(x) = \sum_n p_n(x) e^{\alpha_n x}
$$

the evaluation can be speeded up significantly.

The naive scaling of the evaluation is $\sim \mathcal{O}(n^3)$ but with the caching it should be closer to $\mathcal{O}(n^2)$.

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def integrate_exp(n, s):
    """ Return the polynomial part P_{n,s}(x) of
    P_{n,s}(x) * e^{s x} = \int x^n e^{s x} dx. 
    
     The lru_cache gives a significant speedup, 
     avoiding re-evaluations for previous (n,s). """
    
    if s == 0:
        return sp.Poly(x**n, x).integrate()
    
    t = sp.Rational(1, s)
    l = [ t ]
    for k in range(n - 1, -1, -1):
        t = -t * sp.Rational(k + 1, s)
        l.append(t)
    p = sp.Poly(l, x)
    
    return p
```

```python
from copy import deepcopy
from collections import defaultdict


x, y = sp.symbols(rf"x, y", real=True)


class exponent_polynomial_sum:
    
    def __init__(self):
        self.terms = {0 : sp.Poly([1], x)}
        
    def __zero(self):    
        return defaultdict(sp.Poly([0], x))
    
    def to_expr(self):
        expr = 0
        for key, val in self.terms.items():
            expr += sp.exp(key * x) * sp.Poly(val, x).as_expr()
        return expr.expand()

    def integrate(self):
        res = self.__zero()
        for key, val in self.terms.items():
            for (n, ), coeff in val.all_terms():
                p = coeff * integrate_exp(n, key)
                res[key] += p
                res[0] -= p.coeff_monomial(1)
        self.terms = res
        return self
    
    def exp_mult(self, exp):
        res = {}
        for key, val in self.terms.items():
            res[key + exp] = val
        self.terms = res
        return self
    
    def __str__(self):
        return str(self.to_expr())
    
    def __call__(self, x_val):
        res = 0
        for n, poly in self.terms.items():
            res += sp.exp(n*x_val) * poly(x_val)
        return res
    
    def copy(self):
        res = exponent_polynomial_sum()
        res.terms = deepcopy(self.terms)
        return res
```

# Dimensional scaling

The scaling of $D_n$ with the order $n$ is controlled by the volume of the N-dimensional simplex
$$
V(S^{(\beta)}_{N}) = \frac{\beta^{N}}{ N! }
$$
and the trivial prefactor
$$
D_n \sim \frac{\delta^n}{2} \frac{ \beta^{2n} } { (2n)! }
$$

The rescaled values diagram values $d_n$ defined as
$$
d_n = 
\frac{2}{\delta^n}
\frac{D_n}{V(S^{(\beta)}_{2n})}
=
\frac{2}{\delta^n}
\frac{(2n)!}{\beta^{2n}}
D_n
$$
are of order unity and for $\epsilon = 0$ we have $d_n = 1$ for all $n$.

```python
def evaluate_diagrams_eps(n_max, ϵ, δ, β, digits=512):
    """ Compute D_n for all n = 1, ..., n_{max} using 
    Sympy and dedicated integration formulas and 
    sum of polynomial time exponent class. """
    
    x = sp.symbols(rf"x", real=True)
    
    calc_D = lambda eps, n : (sp.factorial(2*n) / β**(2*n) * eps.copy().exp_mult(-ϵ).integrate()(β)).n(n=digits)
    
    eps = exponent_polynomial_sum().exp_mult(ϵ).integrate()
    D1 = calc_D(eps, 1)
    
    Ds = [D1]    

    for n in tqdm(range(2, n_max+1)):
        eps.exp_mult(ϵ).integrate().exp_mult(-ϵ).integrate()
        Dn = calc_D(eps, n)
        Ds.append(Dn)
            
    Ds = np.array(Ds, dtype=float)
    return Ds
```

\frac{\tau^{2n+1}}{(2n)! (2n + 1)}
$$

```python
ϵ = 1
eps = exponent_polynomial_sum().exp_mult(ϵ).integrate()
print(eps.to_expr())

coeffs_n = []
#ns = np.arange(64)
ns = np.arange(8)

refs = np.array([1/sp.factorial(2*n) for n in np.arange(5)], dtype=float)
print(refs)

for i in ns:
    eps.exp_mult(ϵ).integrate().exp_mult(-ϵ).integrate()
    expr = eps.to_expr()
    print(f'i = {i} ----', )
    n = (2*(i+1)+1)
    m = 8
    expr = expr.series(x, n=n+m) * sp.factorial(2*(i+1))
    expr = (expr * n / x**n).simplify()

    p = sp.Poly(expr.removeO())
    coeffs = np.array([ p.coeff_monomial(x**i) for i in range(0, m) ], dtype=float)
    print(coeffs)
    coeffs_n.append(coeffs)
    
coeffs_n = np.array(coeffs_n)
```

## Empirical scaling for large $n$

We observe that the polynomial coefficients converge towards

$$
\lim_{n\rightarrow \infty} p_n(x) = \sum_{k=0}^\infty \frac{1}{k!} \frac{(-1)^k x^k}{2^k} = e^{-x/2}
$$

$$
I_n(\tau) = \frac{e^{\epsilon x}}{2^n} + p_n(x) e^{-\epsilon x} + q_n(x)...
$$

$$
p_n(x) = C_{n+1} x^{n+1} + ... - C_0
\\
q_n(x) = Q_{n} x^n + ... + C_0 - 1
$$

# Optimized evaluation

For $\epsilon\beta = 1$ the coefficients are integers and the integrations can be cast to matrix vector multiplications.


```python

class exponent_polynomial_sum_numpy:

    def __init__(self, N=1, epsilon=1, build_mats=True):
        
        assert( epsilon == 1 )
        
        self.N = N
        self.epsilon = epsilon
        
        self.dtype = sp.Rational
        self.dtype_int = sp.Integer
        
        self.terms = {0 : np.zeros(1, dtype=self.dtype_int)}
        self.terms[0][0] = 1
        
        if build_mats:
            print(f'--> Build mats N = {N}')
            self.int_mats = {}
            for s in [-2, -1, 0, 1, 2]:
                exp = s * epsilon
                M = np.zeros((N, N+1), dtype=self.dtype)
                for n in range(N):
                    p = integrate_exp(n, exp)
                    new = np.array(p.all_coeffs(), dtype=self.dtype)[::-1]
                    M[n, :len(new)] = new
                self.int_mats[exp] = np.array(M[:, :N]).T
            print('done.')
        
    def __zero(self):    
        return defaultdict(lambda : np.zeros(1, dtype=self.dtype_int))
    
    def __str__(self):
        return str(self.to_expr())
        
    def exp_mult(self, exp):
        res = {}
        for key, val in self.terms.items():
            res[key + exp] = val
        self.terms = res
        return self

    def mult(self, scalar):
        for key, val in self.terms.items():
            self.terms[key] *= scalar
        return self    
    
    def integrate(self):
        res = self.__zero()
        for exp, coeffs in self.terms.items():
            n = len(coeffs)
            eidx = n if exp != 0 else n + 1 # Only exp == 0 increase poly order
            new_coeffs = self.int_mats[exp][:eidx, :n] @ coeffs
            res[0][0] -= new_coeffs[0]

            new_coeffs[:len(res[exp])] += res[exp]
            res[exp] = new_coeffs

        self.terms = res
        return self
    
    def to_expr(self):
        expr = 0
        for e, coeff in self.terms.items():
            expr += sp.exp(e * x) * sp.Poly(coeff[::-1], x).as_expr()
        return expr.expand()
    
    def __call__(self, x_val):
        assert( x_val == 1 )
        res = 0
        max = 0
        for n, coeffs in self.terms.items():
            max = np.max([max, np.max(np.abs(coeffs))])
            res += sp.exp(n*x_val) * np.sum(coeffs)
        res = sp.simplify(res)
        
        dps = np.max([15, sp.log(max, 10).evalf().round()])
        res_num = res.evalf(dps)
        return res_num
    
    def copy(self):
        res = exponent_polynomial_sum_numpy(
            N=self.N, epsilon=self.epsilon, build_mats=False)
        res.terms = deepcopy(self.terms)
        res.N = self.N
        res.int_mats = self.int_mats
        return res
```

```python
   
def evaluate_diagrams_eps_numpy(n_max, ϵ, β, digits=512):
    """ Compute D_n for all n = 1, ..., n_{max} using 
    Sympy and dedicated integration formulas and 
    sum of polynomial time exponent class. """
    
    x = sp.symbols(rf"x", real=True)
    
    calc_D = lambda eps, n : eps.copy().exp_mult(-ϵ).integrate()(β)
    
    eps = exponent_polynomial_sum_numpy(N=n_max, epsilon=ϵ).exp_mult(ϵ).integrate()
    eps.mult(2)
    
    D1 = calc_D(eps, 1)
    
    Ds = [D1]    
    
    for n in tqdm(range(2, n_max+1)):
        eps.exp_mult(ϵ).integrate().exp_mult(-ϵ).integrate()
        eps.mult(2*n).mult(2*n-1)
        Dn = calc_D(eps, n)
        Ds.append(Dn)
            
    Ds = np.array(Ds, dtype=float)
    return Ds    
    
```

# Algebraic evaluation with the incomplete Gamma function

# Evaluation using recursion on $d_n(x)$ with convolution kernel $K(x)$

```python
def exponent_in_legendre(h, beta, N, xi):

    """ Computes the imaginary time Green's function of a non-interacting system
    in Legendre spectral space. 

    The problem is mapped to a block tri-diagonal linear equation system
    solved using a linear scaling on-the-fly LU factorization. The Hamiltonian
    H is diagonalized as a pre-processing step. """

    beta_2 = beta / 2
    if type(h) == float:
        I = 1.
        dtype = float
    else:
        I = sp.Integer(1)
        dtype = sp.Rational

    alpha_inv = np.empty((N), dtype=dtype)
    
    if xi == -1.:
        alpha_inv[0] = I
    else:
        alpha_inv[0] = 1 / (beta_2 * h)
    
    y = np.empty_like(alpha_inv)
    y[0] = -I / 2
    
    # ---------------------------------------------
    # -- FWD
    # ---------------------------------------------

    # -- First step

    q0 = beta_2
    b_1 = q0 * h
    beta_1 = b_1 * alpha_inv[0]
    y[1] = - beta_1 * y[0]

    if xi == 1.:
        alpha_inv[1] = I
    else:
        q1 = beta_2 / 3
        c_1 = q1 * h
        alpha_inv[1] = 1 / (1 + beta_1 * c_1) 

    q0 = beta_2 / 3
        
    # -- Intermediate steps

    for i in range(2, N - 1):

        q1 = beta_2 / (2*i + 1)

        b_i = q0 * h
        c_i = q1 * h
        
        beta_i = b_i * alpha_inv[i - 1]
        alpha_inv[i] = 1 / (1 + beta_i * c_i) 
        y[i] = - beta_i * y[i - 1]

        q0 = q1

    # -- Last step: i = N - 1

    b_i = q0 * h
    beta_i = b_i * alpha_inv[N - 2]
    y[N - 1] = - beta_i * y[N - 2]

    # ---------------------------------------------
    # -- BWD
    # ---------------------------------------------

    # -- First steps

    x = y # reuse data
    x[-2] = x[-2] * alpha_inv[-2]

    # -- Intermediate steps
    
    for i in range(N-3, 0, -1):
        q0 = beta_2 / (2*i + 3)
        c_i = q0 * h

        x[i] += c_i * x[i + 1]
        x[i] = x[i] * alpha_inv[i]

    # -- Last step

    if xi == -1:
        q0 = beta_2 / 3
        c_0 = q0 * h
        x[0] += c_0 * x[1]

    x[0] = x[0] * alpha_inv[0]

    return x
```

```python
def spectral_convolution_matrix_partial_scalar_python(A_n, s):

    N = A_n.shape[0]
    B = np.zeros((N, N), dtype=A_n.dtype)
    a_n = A_n

    # -- column 0
    B[0, 0] = s * a_n[0] - a_n[1] / 3
    for k in range(1, N - 1):
        B[k, 0] = a_n[k - 1] / (2*k - 1) - a_n[k + 1] / (2*k + 3)
    B[N - 1, 0] = a_n[N - 2] / (2*(N - 1) - 1)

    # -- column 1
    B[0, 1] = -s * B[1, 0] / 3
    for k in range(1, N - 1):
        B[k, 1] = B[k - 1, 0] / (2*k - 1) \
            - s * B[k, 0] - B[k + 1, 0] / (2*k + 3)
    B[N - 1, 1] = B[N - 2, 0] / (2*(N - 1) - 1) - B[N - 1, 0]

    # -- Recurse columns
    for n in range(1, N - 1):
        for k in range(n + 1, N - 1):
            B[k, n + 1] = -(2*n + 1) * B[k + 1, n] / (2*k + 3) \
                + (2*n + 1) * B[k - 1, n] / (2*k - 1) + B[k, n - 1]
        k = N - 1
        B[N - 1, n + 1] = (2*n + 1) * B[k - 1, n] / (2*k - 1) + B[k, n - 1]

    # -- Transpose tril to triu        
    for k in range(N - 1):
        for n in range(N):
            B[k, n] = (-1)**(n + k) * (2*k + 1) * B[n, k] / (2*n + 1)

    return B
```
